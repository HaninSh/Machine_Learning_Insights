## Dealing with null values:
1. removing null values
2. Mean imputation
3. KNN imputation


### #1 Removing null values >> removing any rows with null values

```python

cleaned_california_housing = california_housing.dropna(how= 'any')
cleaned_california_housing
```

```python
cleaned_california_housing.isnull().sum()
```

### #2 Mean Imputation

```python
# replacing the null values in 'total_bedrooms' column with the mean value
imputed_total_bedrooms = pd.DataFrame(california_housing['total_bedrooms'].fillna(california_housing['total_bedrooms'].mean()))
imputed_total_bedrooms
```

```python
#checking that the column has been imputed and no null values anymore

imputed_total_bedrooms.isnull().sum()
```

```python
#adding the imputed column to the data frame

imputed_california_housing = california_housing
imputed_california_housing['total_bedrooms'] = imputed_total_bedrooms
imputed_california_housing.isnull().sum()
```

### #3 KNN imputation

```python
# Fisrt we must use the categorically converted data frame
#Therefore, we will drop the non categorical column 'ocean_proximity' and depend on the numerical column 'label_encoded_ocean_proximity'
categorically_converted_california_housing = california_housing.drop( columns= 'ocean_proximity')
categorically_converted_california_housing
```

```python
from sklearn.impute import KNNImputer

# Initialize the KNN Imputer
knn_imputer = KNNImputer(n_neighbors=2)
imputed_data = knn_imputer.fit_transform(categorically_converted_california_housing)
imputed_california_housing = pd.DataFrame(imputed_data)
imputed_california_housing
```

```python
imputed_california_housing.isnull().sum()
```

Matrix Transformations

```python
# first we need to make a 2D array to perform matrix operations

matrix = sliced_array.reshape(5,4)
matrix
```

```python
# Transposition
transpose_matrix = np.transpose(matrix)
transpose_matrix
```

```python
# Inversion
# inverse matrix can be calculated only for square matrices

temp_array = array[0:16]                          # slice original array to obtain a square matrix
square_matrix = temp_array.reshape(4, 4)          # convert the 1D array into 2D array matrix
inverse_matrix = np.linalg.inv(square_matrix)
print (inverse_matrix)
```

```python
# Multiplication
# first we need to have two matrices to perform multiplication operation

array_a = array[0:10]
array_b = array[10:20]

matrix_a = array_a.reshape(5, 2)
matrix_b = array_b.reshape(2, 5)

product = np.dot(matrix_a, matrix_b)
print(product)
```

# Part 4: Feature Engineering

Create new feature : capita room availability

```python
# calculating room availability = (total_rooms / population)

totalRooms = np.array(california_housing['total_rooms'])
population = np.array(california_housing['population'])

room_availability = totalRooms / population
california_housing['room_availability'] = room_availability

print(california_housing[['total_rooms', 'population', 'room_availability']].head())
```

Create new feature : income to price ratio

```python
# income to price ratio = (median_income / median_house_value).

median_income = np.array(california_housing['median_income'])
median_house_value = np.array(california_housing['median_house_value'])

income_price_ratio = median_income / median_house_value
california_housing['income_price_ratio'] = income_price_ratio

print(california_housing[['median_income', 'median_house_value', 'income_price_ratio']].head())
```
